**1. Generate a normal distribution (mean = 7, standard deviation = 2) , for three different sample sizes 100,1000,100,000. Design a python class to get the following items:**

a) Show the density curve for all the three sample sizes (5 points)

b) Explain the findings for all three samples (5 points)

c) Convert the simulated data into a standard normal distribution, and plot for all three different sample sizes. (5 points)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class normal_distribution():
    def __init__(self, mean, std, size):
        """Initialize normal distribution.

        Parameters
        ----------
        param1: size
        param2: mean
        param3: std
        """
        
        self.size = size
        self.mean = mean
        self.std = std
        
        x = np.linspace(mean - 4*std, mean + 4*std, size)
        
        dist = normal_distribution.normal_pdf(mean,std,x)
        
        self.dist = dist
        
        
    def normal_pdf(mean,std,x):
        """Apply the normal distribution formula on the dataset

        Parameters
        ----------
        param1: n
        param2: p
        param3: trails
        """
        sigma2 = std**2
        dist = 1 / np.sqrt(2 * np.pi * sigma2) * np.exp(-(x - mean)**2 / (2 * sigma2))
        return dist
        
    def plotpdf(self):
        """Plot density function.

        Parameters
        ----------
        param1: self
        """
        
        xrange = np.linspace(self.mean - 3*self.std, self.mean + 3*self.std, self.size)
        
        plt.plot(xrange, self.dist)
        plt.show()
        
    def convert_to_standard(self):
        """Covert the distribution to standard normal distribution. And then plot

        Parameters
        ----------
        param1: self
        """
        
        z = (self.dist - self.mean) / self.std
        
        xrange = np.linspace(-4, 4, self.size)
        
        plt.plot(xrange, z)
        plt.show()
```

## Sample Size 100

```python
normal_100 = normal_distribution(7,2, 100)

normal_100.plotpdf()
```

I also found that most of the data are clustered around the center, while the more extreme values on either side of the centre become less rare as the distance from the centre increases. It looks that 68% of the distribution are around range from 5 to 9, and 95% of the data are around range 3 to 11, and 98% of the data are around range 1 to 13.

**3. Binomial Distribution simulation. Generate a random number between 0 and 1. If that number is 0.5 or more, then event it as fake. Otherwise not a fake for 500, 5000, and 500,000 trails.**

Design a python class to get the following items:

a) Show the density curve for all the three sample sizes (5 points)

b) Explain the findings for all three samples (5 points)

```python
from numpy import random
```

```python
class binomial_distribution():
    def __init__(self, p, trails):
        """Initialize binomial distribution.

        Parameters
        ----------
        param1: p
        param2: trails
        """
        #design n as 50
        
        n = 50
        
        self.n = 50
        self.p = p
        self.trails = trails
        
        fake = []
        notfake = []
        
        for i in range(trails):
            train = [np.random.uniform(0,1) for i in range(n)]
            fake.append(len([i for i in train if i >= 0.5]))
            notfake.append(len([i for i in train if i < 0.5]))
            
        fake = np.array(fake)
        notfake = np.array(notfake)
        
        self.fake = fake
        self.notfake = notfake
        
        self.fakedist = binomial_distribution.binomial_pmf(p, trails, fake, 'Fake')
        self.notfakedist = binomial_distribution.binomial_pmf(p, trails, notfake, 'Not Fake')

            
    def factorial(x):
        f = 1
        for i in range (x):
            f = f * (i+1)
        return f
    
    def binomial_coeff(a,b):
        comb = binomial_distribution.factorial(a) / (binomial_distribution.factorial(b) * binomial_distribution.factorial(a - b))
        return comb
            
    def binomial_pmf(p, trails, num, fake_or_not):
        """Apply the binomial distribution formula on the dataset

        Parameters
        ----------
        param1: self
        param2: fake_or_not
        """
        if (fake_or_not == 'Fake'):
            dist = binomial_distribution.binomial_coeff(trails, num) * (p**num) * ((1 - p)**(trails - num))
        elif (fake_or_not == 'Not Fake'):
            dist = binomial_distribution.binomial_coeff(trails, num) * (p**(trails - num)) * ((1 - p)**num)
            
        return dist
    
    def plotpmf(self):
        """Plot probability mass function.

        Parameters
        ----------
        param1: self
        """
        
        plt.plot(self.fake, self.fakedist, color = 'blue')
        plt.title('Fake Cases Binomial Distribution')
        plt.show()
        
        plt.plot(self.notfake, self.notfakedist, color = 'red')
        plt.title('Not Fake Cases Binomial Distribution')
        plt.show()
```
